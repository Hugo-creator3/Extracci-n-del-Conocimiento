```python
datos = {
    "usuario": ["ana", "pepe", "Luis", "rosita", "carolina", "paco"],
    "edad": [14, 22, 35, 40, 62, 70],
    "ciudad": ["cuitlahuac", "yanga", "atoyac", "paraje", "fortin", "cuichapa"], # <-- ¡Arreglado aquí!
    "compras_realizadas": [5, 10, 1, 2, 15, 3],
    "activo": [True, True, False, True, False, True],  
    "fecha_registro": ["2026-01-10", "2026-05-21", "2026-02-10", "2026-03-11", "2026-04-15", "2026-07-25"]
}
```

```python
#Creamos el data fram
dfDatos = pd.DataFrame(datos)
```

```python
dfDatos
```

```python
#Crear el dataFrame de 1000000 de registros
cantidad = 1000000
dfMillon = pd.DataFrame({
    "nombre": [random.choice(dfDatos["usuario"]) for _ in range(cantidad)],
    "edad": [random.randint(18, 80) for _ in range(cantidad)],
    "ciudad": [random.choice(dfDatos["ciudad"]) for _ in range(cantidad)], 
    "compras_realizadas": [random.randint(1, 10) for _ in range(cantidad)],
    "activo": [random.choice(dfDatos["activo"]) for _ in range(cantidad)],
    "fecha_registro": [random.choice(dfDatos["fecha_registro"]) for _ in range(cantidad)]
})
```

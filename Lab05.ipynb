```python
import pandas as pd
import numpy as np
import random
```

```python
from faker import Faker
```

```python
pip install Faker
```

```python
fake = Faker("es_MX")
```

```python
datos = []
for i in range(100):
    datos.append(
        {
            "id_usuario": i + 1,
            "nombre": fake.name(),
            "correo": fake.email(),
            "edad": random.randint(18, 65),  '
            "seguidores": random.randint(50, 5000),
            "plataforma": random.choice(["TikTok", "Instagram", "Youtube"])
        }
    )
```

```python
dfInfluencer=pd.DataFrame(datos)
```

```python
# Graficar
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Configuración del gráfico
plt.figure(figsize=(8, 4))
sns.histplot(dfInfluencer["edad"], bins=10)
plt.title("Distribucion de edades")
plt.show()
```

```python
#Graficamos la cantidad de usuarios por plataforma
plt.figure(figsize=(8,4))
sns.countplot(x="plataforma", data = dfInfluencer)
plt.title("Cantidad de usuarios por plataforma")
plt.show()
```
